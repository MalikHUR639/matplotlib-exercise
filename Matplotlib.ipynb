# MatplotLib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= [1,2,3,4,5,6,7]
y= [35,24,24,45,57,23,67]
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Pta nai')
#alpha controls transperancy(0 to 1)
plt.plot(x,y,'r-o',alpha=1)
```

```python
days= [1,2,3,4,5,6,7]
max_t= [35,24,24,75,57,23,67]
avg_t= [63,22,21,46,59,32,76]
min_t= [3,26,28,15,24,23,12]
```

```python

plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Pta nai')
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend(shadow=True)
plt.grid()
```

# Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
cmp=["google",'amzn','fb']
reven=[24,34,33]
profit=[7,5,29]
```

```python
xpos=np.arange(len(cmp))
```

```python
plt.xticks(xpos,cmp)
plt.ylabel("Revenue(B)")
plt.title("Stocks")
plt.bar(xpos,reven,label='Revenue')
plt.bar(xpos,profit,label='Profit')
plt.legend(shadow=True)
```

# side bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
cmp=["google",'amzn','fb']
reven=[24,34,33]
profit=[7,5,29]
```

```python
xpos=np.arange(len(cmp))
```

```python
plt.xticks(xpos,cmp)
plt.ylabel("Revenue(B)")
plt.title("Stocks")
plt.ylim(0,40)
plt.bar(xpos-0.2,reven,width=0.4,label='Revenue')
plt.bar(xpos,profit,width=0.4,align='edge',label='Profit')
plt.legend(shadow=True)
```

```python
plt.yticks(xpos,cmp)
plt.xlabel("Revenue(B)")
plt.title("Stocks")
plt.barh(xpos,reven,label='Revenue')
plt.barh(xpos,profit,label='Profit')
plt.legend(shadow=True)
# plt.grid()
```

# Pie chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
pie_vals=np.random.randint(0,100,5)
pie_name=["food",'med','other','fees','phone']
```

```python
plt.axis("equal")
plt.pie(pie_vals,labels=pie_name,shadow=True,radius=1.5,autopct='%0.2f%%')#outopct is used to show percentages in chart
plt.show()
```

```python
plt.axis("equal")
plt.pie(pie_vals,labels=pie_name,shadow=True,radius=1.5,autopct='%0.2f%%',explode=[0.1,0.2,0.1,0.1,0.1])#explode used when 1 piece have to come out
plt.savefig("Chart.png")
plt.show()
```
